```python
import pandas as pd
import numpy as np
from glob import glob
import cv2
import matplotlib.pylab as plt
```

# Reading Images

```python
image_file = glob("test.jpg")
```

```python
img = cv2.imread(image_file[0])
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
flat_img = img.reshape((-1,3))
```

```python
flat_img
```

```python
flat_img = flat_img/255
```

```python
flat_img
```

```python
from sklearn.cluster import KMeans
```

```python
k_means = KMeans(n_clusters=3)
```

```python
k_means.fit(flat_img)
```

```python
new_img_flat = k_means.cluster_centers_[k_means.labels_]
```

```python
final_img = new_img_flat.reshape(img.shape)
```

```python
plt.imshow(final_img)
```

```python
plt.imsave("vintage_photo.jpg", final_img)
print("Saved!")
```
